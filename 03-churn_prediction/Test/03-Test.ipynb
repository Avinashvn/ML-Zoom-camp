```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
# Since we cannot see all the columns in one go, we transpose the dataset to see all the colunms

df.head().T
```

```python
# Now lets clean the dataset
# begins with the columns

df.columns = df.columns.str.lower().str.replace(' ', '_')

categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)

for c in categorical_columns:
    df[c] = df[c].str.lower().str.replace(' ', '_')
```

```python
df.head().T
```

```python
df.dtypes
```

```python
tc = pd.to_numeric(df.totalcharges, errors = 'coerce')

df.totalcharges = pd.to_numeric(df.totalcharges, errors = 'coerce')

df.totalcharges = df.totalcharges.fillna(0)
```

```python
df.churn.head()
```

```python
df.churn = (df.churn == 'yes').astype(int)
```

### 3.2 Setting up the Validation Framework

Prepare the train/validation/test split with Scikit-Learn

```python
# Lets split the data from sklearn.model.model_selection
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size = 0.2, random_state = 1)
```

```python
df_train , df_val = train_test_split(df_full_train, test_size = 0.25, random_state = 1)
```

```python
df_train = df_train.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
```

```python
y_train = df_train.churn.values
y_val = df_val.churn.values
y_test = df_test.churn.values
```

```python
del df_train['churn']
del df_val['churn']
del df_test['churn']
```
