```python
# Now we need to find a way to do this calculation to all the features without manually doing.
# first lets see an SQL quary to do this and the lets do it in pandas
```

SELECT

    gender,

    AVG(churn),

    AVG(churn) global_churn AS diff,

    AVG(churn) / global_churn AS risk

FROM

    data

GROUP BY

    gender;

```python
# In pandas we cannot do this whole things in one function.
# we need to do things step by step

df_group = df_full_train.groupby('gender').churn.agg(['mean', 'count'])
df_group['diff'] =  df_group['mean'] -  global_churn_rate 
df_group['risk'] = df_group['mean'] / global_churn_rate

# This is how, we find the relevent details for one feature.
# So to do this to all the required feature, we can use a loop
```

```python
# First lets import a library to display the values in the loop

from IPython.display import display
```

```python
for c in categorical : 
    print(c)
    df_group = df_full_train.groupby(c).churn.agg(['mean', 'count'])
    df_group['diff'] =  df_group['mean'] -  global_churn_rate 
    df_group['risk'] = df_group['mean'] / global_churn_rate
    display(df_group)
    print()
    print()
    
```

### 3.5 Feature Importance : Mutual Information

Mutual information - concept from information theory, it tells us how much we can learn about one variable if we know the value of another

```python
from sklearn.metrics import mutual_info_score

mutual_info_score(df_full_train.churn, df_full_train.contract)

# This function takes target and the feature that we need to learn as input
# the order inside the paranthesis doesn't matter
# Higher the score, it has greater impact in churn rate
```

```python
mutual_info_score(df_full_train.partner, df_full_train.churn)
```

```python
mutual_info_score(df_full_train.churn, df_full_train.gender)
```

```python
# Now lets do this to all the features, for that we need to have a function

def mutual_info_churn_score(series):
        return mutual_info_score(series, df_full_train.churn)   
```

```python
mi = df_full_train[categorical].apply(mutual_info_churn_score)
mi.sort_values(ascending = False)
```
